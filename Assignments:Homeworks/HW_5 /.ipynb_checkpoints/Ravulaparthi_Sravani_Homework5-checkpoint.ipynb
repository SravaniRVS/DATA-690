### Problem 1: (5 points)

<span  style="color:purple; font-size:16px">Create a single DataFrame `shootings` by stacking the 2018 and 2019 mass shooting data and reseting the index. As of August 31, 2019, how many mass shootings were there in the US in 2018 and 2019?</span>

````python
# Answer

# Combine the data for 2018 and 2019 shootings into a single DataFrame
shootings = pd.concat([shooting_2018, shooting_2019]).reset_index(drop=True)

print(f"As of August 31, 2019, there were \033[1;31m{shootings.shape[0]}\033[0m mass shootings in the US in 2018 and 2019.")
````

<span  style="color:green; font-size:14px">

+ Used **```.reset_index(drop=True)```** to reset the index of the concatenated dataframe and it drops the old index column
+ And assigned the resulting dataframe to the variable named **```shootings```** 
+ Found out the total number of shootings using **```.shape[0]```** property of a dataframe that returns the number of rows in the dataframe
</span>

### Problem 3: (5 points)

### Problem 5: (5 points)

<span  style="color:purple; font-size:16px">The problem with the merge in the previous problem was that there was extra whitespace in one or both of the DataFrames. We can fix this using the `str.strip()` method. Apply this method to the relevant columns needed for the join, then repeat the merge above. This time you should be successful. (After this exercise, you should have a DataFrame called `shootings` with the following columns: 'Date', 'Dead', 'Injured', 'Total', 'Description', 'City', 'State', 'Abbreviation')

````python
# Answer 

# Remove leading/trailing whitespaces from the relevant columns
shootings['State'] = shootings['State'].str.strip()
abbr['Abbreviation'] = abbr['Abbreviation'].str.strip()

# Merge the dataframes on 'State' column
shootings = pd.merge(shootings, abbr, on='State', how='left')
shootings
````

````python
shootings['Abbreviation'].isnull().sum()
````

  + and the leading and trailing whitespace from the Abbreviation column in the abbr dataframe 

+ Used **```.merge()```** method to merge the two dataframes **(shootings and abbr)** on the State column and using a left join (the default)

+ And saved it back to, **```shootings```**, which contains all the columns from both the dataframes (shootings and abbr), with any missing values in the **Abbreviation** column filled with **NaN**

### Problem 7: (5 points)

<span  style="color:purple; font-size:16px">Extract the `State` column from `shootings` as a DataFrame. Identify all states/US territories that did not have a mass shooting in 2018 or 2019 (as of 8/31/19) using an appropriate join with the `abbr`. Call the resulting table `shooting_free_states`. (Hint: The `indicator` argument of `merge()` could be useful.) How many states/territories are contained in `shooting_free_states`?</span>

````python
# Answer

# Extract the 'State' column from the 'shootings' DataFrame
state_df = shootings[['State']]

# Merge with the 'abbr' DataFrame to identify shooting-free states
shooting_free_states = pd.merge(abbr, state_df, how='outer', on='State', indicator=True)

# Filter out the states that had mass shootings and drop the '_merge' column
shooting_free_states = shooting_free_states[shooting_free_states['_merge'] == 'left_only']

# Print the result
print(f"The number of states/territories that did not have a mass shooting in 2018 or 2019 (as of 8/31/19) is: \033[1;31m{len(shooting_free_states)}\033[0m")
````

<span  style="color:red; font-size:16px"> **Explanation:**</span>

<span  style="color:green; font-size:14px"> 

+ Used **```double bracket indexing```** to create a new dataframe, **```state_df, containing only the State column from the shootings DataFrame 
    
+ **```pd.merge(abbr, state_df, how='outer', on='State', indicator=True)```**:
   + merges this new **state_df** dataframe with the **abbr** dataframe using an outer join **(how='outer')** on the **State** column
    
   + The **```indicator=True```** argument adds a column called **```_merge```** to the resulting dataframe, **```shooting_free_states```**, that shows which dataframes the rows came from **```(both, left_only, or right_only)```**
    
+ The third line filters the resulting **shooting_free_states** dataframe to include only rows where **_merge** is equal to **left_only**, indicating that the state did not appear in the shootings dataframe
    
+ The fourth line prints the number of rows in the shooting_free_states dataframe using **```len()```**, which gives the number of states/territories that did not have a mass shooting in 2018 or 2019 (as of 8/31/19)


</span>

### Problem 9: (5 points)

<span  style="color:purple; font-size:16px">Remove the strange characters from the `Dead` and `Injured` columns. Convert the `Dead` and `Injured` columns to integers. Print something to verify that you successfully converted the columns to integers. (Hint: it might help to break this into steps.)</span>

````python
# Answer

# Step 1: Verify dtype before conversion
print(f"Dead column is an \033[1;31m{shootings['Dead'].dtype}\033[0m before conversion.")
print(f"Injured column is an \033[1;31m{shootings['Injured'].dtype}\033[0m before conversion.")

# Step 2: Replace strange characters
shootings['Dead'] = shootings['Dead'].replace('\D', '', regex=True)
shootings['Injured'] = shootings['Injured'].replace('\D', '', regex=True)

# Step 3: Convert to integers
shootings['Dead'] = shootings['Dead'].astype(int)
shootings['Injured'] = shootings['Injured'].astype(int)

# Step 4: Verify conversion
print(f"\nDead column has been converted to \033[1;31m{shootings['Dead'].dtype}\033[0m sucessfully.")
print(f"Injured column has been converted to \033[1;31m{shootings['Injured'].dtype}\033[0m sucessfully.")
````

<span  style="color:red; font-size:16px"> **Explanation:**</span>

<span  style="color:green; font-size:14px">

Step 1: The dtype of **Dead** and **Injured** columns are printed before conversion to verify if the **dtype** is currently **not int**
    
Step 2: The non-digit characters are removed from the **Dead** and **Injured** columns using the **```.replace()```** method and the regex pattern **```\D```** (replaces all non-digit characters (\D) in the string with an empty string)
    
Step 3: The **Dead** and **Injured** columns are converted to integers using the **```.astype()```** method with the **int** parameter
    
Step 4: The dtype of **Dead** and **Injured** columns are printed after conversion to verify if the conversion is sucessful or not

### Problem 10: (5 points)

<span  style="color:purple; font-size:16px">Create a horizontal bar graph showing the number of shootings per state. Make the `figsize = (15,10)`. What are the top 3 states in terms of the number of mass shootings in 2018/2019?</span>

````python
# Answer

import matplotlib.pyplot as plt

# Count the number of shootings in each state
shootings_per_state = shootings['State'].value_counts()

# Create a new figure with a given size
plt.figure(figsize=(15, 10))  

# Create a horizontal bar plot of the shootings_per_state data
shootings_per_state.plot(kind='barh', color = 'orange')

# Set X-label, Y-label and Title for the Plot
plt.xlabel('Number of Shootings', fontweight = 'bold')
plt.ylabel('State', fontweight = 'bold')
plt.title('Number of Mass Shootings per State (2018-2019)', fontweight = 'bold', color = 'r')

# Display the plot
plt.show()
````
